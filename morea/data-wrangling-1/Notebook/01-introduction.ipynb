```python
import numpy as np

mean = 0          # mean (center) of the distribution
std_dev = 1       # standard deviation (spread or width) of the distribution
num_samples = 1000  # number of samples to generate

samples = np.random.normal(mean, std_dev, num_samples)

print(samples[:10])
```
